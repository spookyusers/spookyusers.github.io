## periodic table

### groups and periods
groups - columns, 1-8 , skip transition metals
period - rows, 1-7
example: C = group4 period2

group1 = alkaline metals
group2 = alkaline earth metals (2+)
group6 = chalogens
group7 = halogens, charge = (1-)
group8 = noble gases

yellow = metals
purple = metaloids (memorize these)
green = nonmetal

elements in the same group have similar chemical properties


### cations anions
cation:
- loses electrons, positive charge
- metals generally make cations
anions:
- gains electrons, negative charge
- non-metals generally

### properties
metals:
- shiny
- conductive (electricity, heat)
- malleable
- mostly solid
- ductile

## chemical bonds

### ionic bonds:
- ionic compounds
- metal + nonmetal
- positive charge + negative charge

### covalent bonds:
- molecular compounds
- between non-metals

### Making ionic compounds
Use chris cross method, top right number goes bottom left, top left to bottom right.

## Nomenclature

metal + nonmetal:
- metal stays the same, nonmetal changes to ending in 'ide'

transition metal + non-metal:
- same but roman numeral after metal to indicate charge

polyatomic ions:
- memorize
- 'name it like you see it'
- metal + polyatomic ion

molecular compounds:

Example: CO_2 = Carbon DiOxide

1 = mono (do not use for first element)
2 = di
3 = tri
4 = tetra
5 = penta
6 = hex
7 = hect
8 = oct
9 = nona
10 = deca

## Measurements

### Moles
number of moles - grams/molar mass
grams to moles: multiply by avogadro's number
moles to grams: divide by avogadro's number

Example: 
23 grams NaOH = mass / sum of molecular mass

1 Na = 23
1 O = 16
1 H = 1
23 + 16 + 1 = 40g/mol
23g/40g/mol = 0.575mol

### Calculating atomic mass of isotope

- change percent to decimal
- multiply decimal by average mass
- add both numbers
